아래는 데이터 A,B,C를 순환신경망으로 다룰때를 나타냈는데 보이는 바와 같이 A를 먼저 처리하여 출력 Oa를 얻고, 그 다음 B를 처리할때 이전에 얻었던 출력 Oa를 같이 사용하여 새로운 출력 Ob를 얻는다. 마찬가지로 C를 처리할 때는 이전에 얻었던 출력 Ob를 같이 사용하여 새로운 출력 Oc를 얻게된다.
